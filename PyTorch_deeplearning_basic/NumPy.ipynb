##配列の形状変換

```python
g=np.arange(9).reshape((3,3))
print('g=',g)
```

```python
g=g.reshape((1,-1))
print(g)
```

```python
#要素がゼロの配列
zeros=np.zeros((3,4))
print('zeros=',zeros)
```

```python
#要素が1の配列
h=np.ones((3,3))
print('h=',h)
```

```python
#未初期化の配列の生成
p=np.empty((2,3))
print('p=',p)
```

```python
#matrixで2次元配列の作成
k=np.matrix([[0,1,2],[3,4,5],[6,7,8]])
print('k=',k)
```

```python
l=np.matrix([[1,1,1],[1,1,1],[1,1,1]])
print('l=',l)

m=np.dot(k,l)
print('m=',m)
print(m.shape)
```
