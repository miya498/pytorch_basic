#簡単なグラフ作成

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#簡単なグラフ描画
plt.plot([1,2,3,4])
plt.show()
```

```python
#配列の要素を逆順
plt.plot([4,3,2,1])
plt.show()
```

```python
#x,yを設定して折れ線グラフ
x=np.array([1,2,3,4,5])
y=np.array([100,200,300,400,500])

plt.plot(x,y)
plt.show()
```

```python
volume=30

x=np.arange(volume)
y=[x_i+np.random.randn(1) for x_i in x]

#polyfitで近似直線の
a,b=np.polyfit(x,y,1)

#x,yが点の座標で'o'で点を表示する。
#a*np.arange(volume)+b,'-'で線を引く
plt.plot(x,y,'o',np.arange(volume),a*np.arange(volume)+b,'-')

```

```python
#0と10の間を100分割
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.show()
```

#グラフの要素に名前を設定する

```python
plt.plot(x,np.arctan(x))
#グラフのタイトル
plt.title('arctan')
plt.xlabel('label for X-axis')
plt.ylabel('label for Y-axis')
plt.show()
```

##グラフのグリッドを表示

```python
plt.plot(x,np.sin(x))

plt.title('sin')
plt.xlabel('label for X-axis')
plt.ylabel('label for Y-axis')
plt.grid(False) #デフォルトはTrue
plt.show()
```

##グラフのメモリの設定

```python
x=np.linspace(0,2*np.pi)
plt.plot(x,np.sin(x))

plt.title('sin')
plt.xlabel('label for X-axis')
plt.ylabel('label for Y-axis')
plt.grid(False) #デフォルトはTrue

position=[0,np.pi/2,np.pi,np.pi*3/2,np.pi*2]
labels=['0','90','180','270','360']
plt.xticks(position,labels)
plt.show()
```

##グラフのサイズ

```python
plt.figure(figsize=(4,4))

x=np.linspace(0,2*np.pi)
plt.plot(x,np.sin(x))
plt.grid(True)

plt.show()
```

##散布図

```python
x=np.random.rand(100)
y=np.random.rand(100)

plt.scatter(x,y)
plt.show()
```

```python
#sは点のサイズ,cは点の色,alphaは点の透明度,linewidthsは線の太さ,edgecolorは点の周囲の色
plt.scatter(x,y,s=600,c="pink",alpha=0.5,linewidths=2,edgecolors="red")
plt.show()
```

#複数のグラフの表示

```python
import matplotlib.pyplot as plt

#2行1列の1番目に表示
plt.subplot(211)
plt.scatter(x,y,s=600,c="pink",alpha=0.5,linewidths=2,edgecolors="red")

#2行1列の2番目に表示
plt.subplot(212)
plt.scatter(x,y,s=300,c=y,cmap="Greens")

plt.show()
```

```python
import matplotlib.pyplot as plt

#1行2列の1番目に表示
plt.subplot(121)
plt.scatter(x,y,s=600,c="pink",alpha=0.5,linewidths=2,edgecolors="red")

#1行2列の2番目に表示
plt.subplot(122)
plt.scatter(x,y,s=300,c=y,cmap="Greens")

plt.show()
```

#3次元散布図

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(0)
random_x=np.random.randn(100)
random_y=np.random.randn(100)
random_z=np.random.randn(100)

fig=plt.figure(figsize=(8,8))

ax=fig.add_subplot(1,1,1,projection="3d")

x=np.ravel(random_x)
y=np.ravel(random_y)
z=np.ravel(random_z)

ax.scatter3D(x,y,z)
plt.show()
```

##色とマーカーを変える

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(0)
random_x=np.random.randn(100)
random_y=np.random.randn(100)
random_z=np.random.randn(100)

fig=plt.figure(figsize=(8,8))

ax=fig.add_subplot(1,1,1,projection="3d")

x=np.ravel(random_x)
y=np.ravel(random_y)
z=np.ravel(random_z)

c="r"
m="^"
s=300

ax.scatter3D(x,y,z,s=s,c=c,marker=m)
plt.show()
```
